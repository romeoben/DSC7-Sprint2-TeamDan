```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
KEYWORD1 = "TIKTOK DANCE"
KEYWORD2 = "TIKTOK OPM"
KEYWORD3 = "TIKTOK PHILIPPINES"
KEYWORD4 = "_TIKTOK"

playlist1_df = pd.read_csv('data/'+KEYWORD1+'_playlist_data.csv')
playlist2_df = pd.read_csv('data/'+KEYWORD2+'_playlist_data.csv')
playlist3_df = pd.read_csv('data/'+KEYWORD3+'_playlist_data.csv')
playlist4_df = pd.read_csv('data/'+KEYWORD4+'_playlist_data.csv')
```

```python
playlist4_df.shape
```

```python
tracks1_df = pd.read_csv('data/'+KEYWORD1+'_playlist_tracks_data.csv')\
                .merge(pd.read_csv('data/'+KEYWORD1+'_playlist_tracks.csv')[['track_id','playlist_id','playlist_name']],\
                      on='track_id',how='left')
tracks1_df['duration_mins']=tracks1_df['duration']/60000
tracks1_df['genre']=KEYWORD1
#--------------------------
tracks2_df = pd.read_csv('data/'+KEYWORD2+'_playlist_tracks_data.csv')\
                .merge(pd.read_csv('data/'+KEYWORD2+'_playlist_tracks.csv')[['track_id','playlist_id','playlist_name']],\
                      on='track_id',how='left')
tracks2_df['duration_mins']=tracks2_df['duration']/60000
tracks2_df['genre']=KEYWORD2
#--------------------------
tracks3_df = pd.read_csv('data/'+KEYWORD3+'_playlist_tracks_data.csv')\
                .merge(pd.read_csv('data/'+KEYWORD1+'_playlist_tracks.csv')[['track_id','playlist_id','playlist_name']],\
                      on='track_id',how='left')
tracks3_df['duration_mins']=tracks3_df['duration']/60000
tracks3_df['genre']=KEYWORD3
#--------------------------
tracks4_df = pd.read_csv('data/'+KEYWORD4+'_playlist_tracks_data.csv')\
                .merge(pd.read_csv('data/'+KEYWORD4+'_playlist_tracks.csv')[['track_id','playlist_id','playlist_name']],\
                      on='track_id',how='left')
tracks4_df['duration_mins']=tracks4_df['duration']/60000
tracks4_df['genre']=KEYWORD4
```

```python
charts = pd.read_csv("data/spotify_daily_charts.csv")
```

```python
charts["track_id"].nunique()
```

```python
tiktok_artists = tracks4_df[tracks4_df["track_id"].isin(charts.track_id)]
```

```python
tiktok_artists["track_id"].nunique()
```

```python
#General Tiktok vs. Tiktok PH

for col in ['danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
       'liveness', 'valence', 'tempo']:
    fig = plt.figure()
    ax= fig.add_subplot(111)
    
    #sns.distplot(tracks1_df[col], ax=ax, label= KEYWORD1)
    #sns.distplot(tracks2_df[col], ax=ax, label= KEYWORD2)
    sns.distplot(tracks3_df[col], ax=ax, label= KEYWORD3)
    sns.distplot(tracks4_df[col], ax=ax, label= KEYWORD4)
    #sns.distplot(tiktok_artists[col], ax=ax, label= "Top 200")
    plt.title("%s vs %s: %s " % (KEYWORD3,KEYWORD4,col))
    plt.ylabel('Frequency')
    plt.legend(frameon=False)
    plt.show()
```

Insights: Filipino TikTok songs follow the trends of general TikTok songs

```python
#TikTok general vs. tiktok ph vs. top 200
#General Tiktok vs. Tiktok PH

for col in ['danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
       'liveness', 'valence', 'tempo']:
    fig = plt.figure()
    ax= fig.add_subplot(111)
    
    #sns.distplot(tracks1_df[col], ax=ax, label= KEYWORD1)
    #sns.distplot(tracks2_df[col], ax=ax, label= KEYWORD2)
    sns.distplot(tracks3_df[col], ax=ax, label= KEYWORD3)
    sns.distplot(tracks4_df[col], ax=ax, label= KEYWORD4)
    sns.distplot(tiktok_artists[col], ax=ax, label= "Top 200")
    plt.title("%s " % (col))
    plt.ylabel('Frequency')
    plt.legend(frameon=False)
    plt.show()
```
